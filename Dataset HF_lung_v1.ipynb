# Creazione dataset e file .tsv per pretrain - Cicli respiratori

```python
!pip install pydub
```

```python
from pydub import AudioSegment
def split_audio(filepath, t1, t2):
  t1 = t1.split(':')[-1]
  t2 = t2.split(':')[-1]
  t1 = float(t1)
  t2= float(t2)
  t1 = t1 * 1000 #Works in milliseconds
  t2 = t2 * 1000
  newAudio = AudioSegment.from_wav(filepath)
  newAudio = newAudio[t1:t2]
  return newAudio
```

```python
directory = r'/content/train'
df_train = pd.DataFrame(columns=['filename'])
df_train = df_train.append({'filename':directory}, ignore_index=True)
for filename in tqdm(os.listdir(directory)):
    if filename.endswith(".wav"):# or filename.endswith(".png"):
      # try:
      with open(os.path.join(directory, filename.replace('.wav','_label.txt')),'r') as file:
        first_I=True
        inizio = 0
        fine = 0
        for num_split, line in enumerate(file):
          entry = line.split(' ')
          #entry[0]=t1, entry[1]=t2, entry[2]=presence of crackles, entry[3].replace('\n','')=presence of wheezels
          if entry[0]=='I' and first_I:
            inizio = entry[1]
            first_I=False
            continue
          elif entry[0]=='I':
            fine = entry[2]
            first_I=True
          else:
            continue
          audio_tmp = split_audio(os.path.join(directory,filename), inizio, fine)
          filename2 = filename.replace('.wav','_'+str(num_split))
          audio_tmp.export('/content/train_cicli_respiratori/'+filename2+'.wav', format='wav')
          df_train = df_train.append({'filename':filename2}, ignore_index=True)
          inizio=0
          fine=0
          break
      # except:
      #   print('FileNotFoundError '+ filename.replace('.wav','.txt'))
      # break
      
    else:
        continue
```
